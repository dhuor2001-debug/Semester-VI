## What is Data Analysis

Data Analysis is the process of collecting, cleaning, organizing data and coming with a conclusion.

# 1.Load the data


```python
import pandas as pd

df = pd.read_excel("data.xlsx")
```

```python
df.head()
```

```python
#Get Summary
print(df.describe())
```

```python
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
df.head()
```

```python
df.groupby("brand")[["price"]].mean()
```

```python
df.groupby("brand").agg({
    "price": ["mean", "min", "max"],
    "Ram": "first",
    "ROM": "first",
    "processor": "first",
    "warranty": "mean"
})
```

```python
df.corr(numeric_only=True)
```

```python
df.corr(numeric_only=True)["price"].sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df["price"], df["resolution_height"], marker = "o")
plt.title("Price vs resolution_height")
plt.xlabel("Price")
plt.ylabel("Resolution_height")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.bar(df["brand"], df["price"])
plt.title("Price vs Brand")
plt.xlabel("Brand")
plt.ylabel("Price")
plt.show()
```

```python
df.hist(bins=30, figsize=(12, 10))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['brand'].value_counts().head(10).plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.85
)
plt.title("Top 10 Brand Distribution")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_brands = df['brand'].value_counts().head(8).index
sns.boxplot(x='brand', y='price', data=df[df['brand'].isin(top_brands)], order=top_brands)
plt.xticks(rotation=45)
plt.title('Price distribution for top 8 brands')
plt.tight_layout()
plt.show()

sns.pairplot(df[['price', 'spec_rating', 'Ram', 'ROM', 'display_size']], diag_kind='kde')
plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.pairplot(df.select_dtypes(include=['number']))
plt.show()
```
