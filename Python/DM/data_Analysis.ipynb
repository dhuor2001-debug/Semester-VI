2. Get Mean, Median and Mode


````python
import matplotlib.pyplot as plt

# Price distribution
df["price"].hist()
plt.show()
````

````python
df.groupby("brand")[["price"]].mean()
````

````python
df.groupby("brand").agg({
    "price": ["mean", "min", "max"],
    "Ram": "first",
    "ROM": "first",
    "processor": "first",
    "warranty": "mean"
})
````

````python
df["Ram"] = df["Ram"].astype(str).str.extract('(\d+)').astype(int)
df["ROM"] = df["ROM"].astype(str).str.extract('(\d+)').astype(int)
````

````python
for col in ["Ram", "ROM"]:
    df[col] = df[col].astype(str).str.extract('(\d+)').astype(int)
````

````python
df.corr(numeric_only=True)
````

````python
df.corr(numeric_only=True)["price"].sort_values(ascending=False)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
````

Good — this is exactly the kind of thing that *clicks once explained properly*. Let’s break your heatmap in a **simple, intuitive way** 👇

---

# 🧠 1. What this heatmap actually is

👉 It shows **correlation (relationship) between numbers**

* Range: **-1 → 0 → +1**

| Value | Meaning                             |
| ----- | ----------------------------------- |
| +1    | strong positive (increase together) |
| 0     | no relationship                     |
| -1    | strong negative (opposite)          |

---

# 🎯 2. Focus ONLY on one thing first: `price`

Don’t try to read everything at once ❌
👉 Just read the **price row**

From your heatmap:

| Feature           | Correlation with Price | Meaning            |
| ----------------- | ---------------------- | ------------------ |
| Ram               | **0.74**               | 🔥 strong positive |
| resolution_height | **0.60**               | strong             |
| resolution_width  | **0.59**               | strong             |
| spec_rating       | **0.55**               | moderate           |
| display_size      | 0.23                   | weak               |
| warranty          | 0.12                   | very weak          |
| ROM               | **-0.48**              | ⚠️ negative        |

---

# 🔥 3. Translate this into REAL meaning

### ✅ Strong relationships

* **RAM (0.74)**
  👉 More RAM → higher price (makes sense)

* **Resolution (0.6)**
  👉 Better screen → higher price

---

### ⚠️ Interesting one

* **ROM (-0.48)**
  👉 Looks like:

  > More storage → lower price ❗

🚨 This is likely **misleading**
Because:

* dataset is small
* some cheap laptops may have large storage

👉 So don’t blindly trust this

---

### ❌ Weak relationships

* Warranty (0.12) → almost no effect
* Display size (0.23) → small effect

---

# 🎨 4. How to read colors (quick trick)

* 🔥 Light/bright → strong positive
* 🌑 Dark → negative
* 🟫 Neutral → weak

---

# 🧠 5. Most important insight (THIS IS THE KEY)

👉 Correlation ≠ causation

Just because:

```
RAM ↑ → price ↑
```

doesn’t mean RAM *alone* controls price.

👉 It just means they move together.

---

# 🧩 6. Why yours feels confusing

Because you're trying to read:

* all rows
* all columns
* all numbers

❌ Don’t do that

---

# ✅ 7. Correct way to read ANY heatmap

### Step 1:

Pick **target variable**
👉 here: `price`

### Step 2:

Read only that row

### Step 3:

Sort mentally:

* High → important
* Low → ignore

---
